```python
import statsmodels.api as sm
```

```python
physical_attributes = ["speed", "jump_height", "endurance", "strength"]

X = df[physical_attributes]
y = df["player_efficiency"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

R-squared value of 0.013 indicates that only 1.3% of variation in player efficiency is explained by these attributes - Maybe not strong predictors?

Smalle p-value of jump_height(0.001) shows jump_height attribute is statistically significant at 5% confidence level meaning it has significant effect on player efficiency
